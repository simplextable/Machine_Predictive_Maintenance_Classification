```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import datetime as dt
import datetime, pytz
import seaborn as sns
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from matplotlib import pyplot
import yfinance
import investpy
from stockstats import StockDataFrame

import matplotlib.dates as mpl_dates
from mpl_finance import candlestick_ohlc
from scipy.stats import skew, norm
from scipy.signal import argrelextrema
import copy
from sklearn.cluster import KMeans 
import time
from sklearn.cluster import MiniBatchKMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report, precision_score, confusion_matrix, auc, f1_score, recall_score

pd.set_option("display.max_rows",2000)
pd.set_option("display.max_columns",50)
```

```python
df = pd.read_csv("predictive_maintenance.csv") 
```

```python
df = df.drop(["UDI",'Product ID'],axis=1)
df['nf'] = df['Tool wear [min]'] * df['Torque [Nm]']
```

```python
df.head()
```

```python
df_target = df.Target.copy()
df_failure_type = df["Failure Type"].copy()
```

```python
df["Failure Type"] = df["Failure Type"].replace("No Failure",0)
df["Failure Type"] = df["Failure Type"].replace("Tool Wear Failure",1)
df["Failure Type"] = df["Failure Type"].replace("Heat Dissipation Failure",2)
df["Failure Type"] = df["Failure Type"].replace("Power Failure",3)
df["Failure Type"] = df["Failure Type"].replace("Overstrain Failure",4)
df["Failure Type"] = df["Failure Type"].replace("Random Failures",5)
```

```python
df["Type"] = df["Type"].replace("L",0)
df["Type"] = df["Type"].replace("M",1)
df["Type"] = df["Type"].replace("H",2)
```

```python
X_train, X_val, y_train, y_val = train_test_split(df.drop(['Failure Type','Target'],axis=1),
                                                    df['Failure Type'], test_size=0.3, random_state=42)
```

```python
def calculate_cost():
    global total_cost_of_not_being_able_to_predict, total_cost_of_action
    
    #açıklama ekleyelim. değişken isimleri uzun oldu ama kod anlaşılsın diye yaptık.
    
    action_unit_cost = 5
    
    power_failure_unit_cost_of_not_being_able_to_predict = 45
    tool_failure_unit_cost_of_not_being_able_to_predict = 30
    overstrain_failure_unit_cost_of_not_being_able_to_predict = 20
    random_failure_unit_cost_of_not_being_able_to_predict = 15
    heat_dissipation_unit_cost_of_not_being_able_to_predict = 5
    
    #--------------------------------------------------------------------------------------------------------------------------
    
    power_failure_number_of_actions = cm[0,1] + cm[1,1] + cm[2,1] + cm[3,1] + cm[4,1] + cm[5,1]
    tool_failure_number_of_actions = cm[0,2] + cm[1,2] + cm[2,2] + cm[3,2] + cm[4,2] + cm[5,2]
    overstrain_failure_number_of_actions = cm[0,3] + cm[1,3] + cm[2,3] + cm[3,3] + cm[4,3] + cm[5,3]
    random_failure_number_of_actions = cm[0,4] + cm[1,4] + cm[2,4] + cm[3,4] + cm[4,4] + cm[5,4]
    heat_dissipation_number_of_actions = cm[0,5] + cm[1,5] + cm[2,5] + cm[3,5] + cm[4,5] + cm[5,5]

    power_failure_total_cost_of_action = power_failure_number_of_actions * action_unit_cost
    tool_failure_total_cost_of_action = tool_failure_number_of_actions * action_unit_cost
    overstrain_total_cost_of_action = overstrain_failure_number_of_actions * action_unit_cost
    random_failure_total_cost_of_action = random_failure_number_of_actions *action_unit_cost
    heat_dissipation_total_cost_of_action = heat_dissipation_number_of_actions * action_unit_cost
    
    #--------------------------------------------------------------------------------------------------------------------------
    
    power_failure_number_of_not_being_able_to_predict = cm[1,0]+cm[1,2]+cm[1,3]+cm[1,4]+cm[1,5]
    tool_failure_number_of_not_being_able_to_predict = cm[2,0]+cm[2,1]+cm[2,3]+cm[2,4]+cm[2,5]
    overstrain_failure_number_of_not_being_able_to_predict = cm[3,0]+cm[3,1]+cm[3,2]+cm[3,4]+cm[3,5]
    random_failure_number_of_not_being_able_to_predict = cm[4,0]+cm[4,1]+cm[4,2]+cm[4,3]+cm[4,5]
    heat_dissipation_number_of_not_being_able_to_predict = cm[5,0]+cm[5,1]+cm[5,2]+cm[5,3]+cm[5,4]
    
    power_failure_total_cost_of_not_being_able_to_predict = power_failure_number_of_not_being_able_to_predict *  power_failure_unit_cost_of_not_being_able_to_predict
    tool_failure_total_cost_of_not_being_able_to_predict = tool_failure_number_of_not_being_able_to_predict * tool_failure_unit_cost_of_not_being_able_to_predict
    overstrain_total_cost_of_not_being_able_to_predict = overstrain_failure_number_of_not_being_able_to_predict * overstrain_failure_unit_cost_of_not_being_able_to_predict
    random_failure_total_cost_of_not_being_able_to_predict = random_failure_number_of_not_being_able_to_predict * random_failure_unit_cost_of_not_being_able_to_predict
    heat_dissipation_total_cost_of_not_being_able_to_predict = heat_dissipation_number_of_not_being_able_to_predict * heat_dissipation_unit_cost_of_not_being_able_to_predict
    
    #--------------------------------------------------------------------------------------------------------------------------

    
    
    total_cost_of_not_being_able_to_predict = (power_failure_total_cost_of_not_being_able_to_predict + tool_failure_total_cost_of_not_being_able_to_predict + overstrain_total_cost_of_not_being_able_to_predict + 
                                              random_failure_total_cost_of_not_being_able_to_predict + heat_dissipation_total_cost_of_not_being_able_to_predict)
        
    total_cost_of_action = power_failure_total_cost_of_action + tool_failure_total_cost_of_action + overstrain_total_cost_of_action + random_failure_total_cost_of_action + heat_dissipation_total_cost_of_action
    
    cost_of_model =  total_cost_of_not_being_able_to_predict + total_cost_of_action
    return cost_of_model
```

```python
classifier=[]
cost = []
```

```python
#Bagging
from sklearn.ensemble import BaggingClassifier
bc = BaggingClassifier()
bc.fit(X_train, y_train)
y_pred = bc.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('BaggingClassifier')
cost.append(cost_level)

#MultiLayerPerceptron
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier()
mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('MLPClassifier')
cost.append(cost_level)

#GBC
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('GradientBoostingClassifier')
cost.append(cost_level)

#ADA
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(X_train, y_train)
y_pred = ada.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('AdaBoostClassifier')
cost.append(cost_level)

#RFC
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('RandomForestClassifier')
cost.append(cost_level)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('LogisticRegression')
cost.append(cost_level)

#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('KNeighborsClassifier')
cost.append(cost_level)

#SVM
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
cost_level = calculate_cost()
classifier.append('Support Vector Machine')
cost.append(cost_level)
```

```python
classifiers = pd.DataFrame({'Classifier':classifier,'Cost':cost})
```
